```python
# Criando nosso dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
# Qual foi a receita total?
df["Valor Venda"].sum()
```

```python
# Qual o custo total?
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"]) # Criando a coluna de custo total
```

```python
df.head()
```

```python
# Qual o custo total?
df["Custo"].sum()
```

```python
# Agora que temos a receita e o custo e o total, podemos achar o lucro total
# Vamos criar uma coluna de lucro que será receita - custo
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head()
```

```python
# Total lucro?
round(df["Lucro"].sum(),2)
```

```python
# Criando uma coluna com total de dias para enviar o produto
df["Tempo_Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
# Verificando o tipo da coluna Tempo-Envio
df["Tempo_Envio"].dtypes
```

```python
# Media de tempo de envio para cada marca
round(df.groupby("Marca")["Tempo_Envio"].mean(),2)
```

```python
# verificando se tem algum dado nulo
df.isnull().sum()
```

```python
# Agrupando por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
# Transformando os números float com as divisoes
pd.options.display.float_format = "{:20,.2f}".format
```

```python
# Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
# Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
# Gráfico total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(color = "chocolate")
plt.title("Total de Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
# Gráfico de lucro por ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(color = "orange")
plt.title("Lucro x Ano")
plt.xlabel("Mês")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
# Selecionando as vendas de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.head()
```

```python
# Grafico de lucro por mes
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot()
plt.title("Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
# Grafico por marca por lucro
df_2009.groupby("Marca")["Lucro"].sum().sort_values(ascending=False).plot.bar()
plt.title("Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Receita")
plt.xticks(rotation='horizontal'); # Eixo X na horizontal
```

```python
df_2009.groupby("Classe")["Lucro"].sum().sort_values(ascending=False).plot.bar(color = "Green")
plt.title("Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Receita")
plt.xticks(rotation = 'horizontal');
```
